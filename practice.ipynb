```python
def model(z,t):
    x,y = z
    dxdt = 5*x-3*y
    dydt = -6*x +2*y
    return dxdt,dydt
z0=[1,2]
t = np.linspace(1,2,100)

# solve ODE
z = odeint(model,z0,t)

x=z[:,0]
y=z[:,1]
x,y
```
